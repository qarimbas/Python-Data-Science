```
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
c = np.cos(x)
s = np.sin(x)

#blue color with a continious line of width 1 (px)
plt.plot(x, c, color="blue", linewidth=1.0, linestyle=":", label="cosine")

#green color with a continious line of width 1 (px)
plt.plot(x, s, color="green", linewidth=1.0, linestyle="-.", label="sine")

plt.xlim(-4.0, 4.0)
plt.ylim(-1.0, 1.0)

#set x ticks
#plt.xticks(-np.linspace(-4,4,5,endpoint=True))
#plt.xticks((-np.pi, -np.pi/2, 0, np.pi/2, np.pi))
plt.xticks((-np.pi, -np.pi/2, 0, np.pi/2, np.pi), (r'$-\pi$', r'$-\pi/2$',r'$0$',r'$\pi/2$',r'$\pi$',))

#set y ticks
plt.yticks(np.linspace(-1,1,5,endpoint=True))

plt.legend()

plt.title("sine - cosine")

plt.show()
```

```
#Scatter Plot

data= {
        "a": np.arange(50),
        "c": np.random.randint(0, 50, 50),
        "d": np.abs(np.random.randn(50)) * 100
      }

#adding a new key
data["b"] = data["a"] + 10 * np.random.randn(50)

plt.scatter(data["a"], data["b"], c=data["c"], s=data["d"]) #setting x and y axes from the data
                                                            # c = color coding, s = size of each dot
plt.xlabel("a")
plt.ylabel("b")
plt.show()
```

```
#Histograms

x = np.random.normal(10, 5, 1000) #normal dist. with a mean set to 10 and standart deviation to 5

plt.hist(x, bins=10)
plt.show()
```

```
#Bar Charts

x = ['Nuclear', 'Hydro', 'Gas', 'Oil', 'Coal', 'Biofuel']
energy = np.random.randint(10, 100, len(x))

plt.bar(x, energy, color='green')
plt.xlabel("Energy Source")
plt.ylabel("Energy Output (GJ)")
plt.title("Energy output from various fuel resources")

plt.show()
```

```
data = { 'apples': 10, 'oranges': 15, 'lemons': 5, 'limes': 20 }
names = list(data.keys())
values = list(data.values())

fig, axs = plt.subplots(1, 3, figsize=(10, 3), sharey=True)

axs[0].plot(names, values)
axs[0].scatter(names, values)
axs[0].bar(names, values)
```

```
#Widgets

from ipywidgets import interact

x = np.linspace(-np.pi, np.pi, 256, endpoint=True)

def plotsin(f): # parameter will be updated with the slider
      plt.plot(x, np.sin(x*f))
      plt.show()

interact(plotsin, f=(1, 10, 1))  #start, end, stride
```

```
n=10
values = np.random.uniform(10, 100, n)
letters = [chr(i) for i in range(ord("A"), ord("Z") + 1)]
categories = np.random.choice(letters, n, replace = False)
categories
```

```
categories = categories[values.argsort()]
values = np.sort(values)

plt.barh(categories, values)

for value, category in zip(values, categories):
  plt.text(value, category, f"{value:.2f}")

plt.xlim((0, values.max() + 10))
plt.title("Horizontal Bar with Text Labels")
plt.show()
```
